### Load the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

csv_file_path = 'fraudTrain.csv'
df = pd.read_csv(csv_file_path)
df.head(10)
```

```python
# Check for missing values
print(df.isnull().sum())

# Check for duplicate records
print("Number of duplicates:", df.duplicated().sum())
```

### Check for class imbalance

```python
print(df['is_fraud'].value_counts())
```

```python
from sklearn.utils import resample
# Separate majority and minority classes
df_majority = df[df.is_fraud == 0]
df_minority = df[df.is_fraud == 1]

print("Before resampling:")
print(df['is_fraud'].value_counts())


df_majority_undersampled = resample(df_majority,
                                    replace=False,
                                    n_samples=20000,
                                    random_state=42)


df_balanced = pd.concat([df_majority_undersampled, df_minority])


df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

print("After resampling:")
print(df_balanced['is_fraud'].value_counts())
```

### Feature importance

```python

from sklearn.ensemble import RandomForestClassifier

datetime_columns = ['trans_date_trans_time']

df_balanced = df_balanced.drop(['Unnamed: 0', 'trans_num', 'cc_num', 'dob', 'first', 'last', 'street', 'merchant'], axis=1)

for column in datetime_columns:
    df_balanced[column] = pd.to_datetime(df_balanced[column], errors='coerce')

# Convert the Timestamps to numerical values
for column in datetime_columns:
    df_balanced[column + '_year'] = df_balanced[column].dt.year
    df_balanced[column + '_month'] = df_balanced[column].dt.month
    df_balanced[column + '_day'] = df_balanced[column].dt.day
    df_balanced[column + '_hour'] = df_balanced[column].dt.hour
    df_balanced[column + '_minute'] = df_balanced[column].dt.minute
    df_balanced[column + '_second'] = df_balanced[column].dt.second

df_balanced = df_balanced.drop(datetime_columns, axis=1)

df_balanced = df_balanced.apply(pd.to_numeric, errors='coerce')

df_balanced = df_balanced.dropna(axis=1)

# Split the features and target variable
X = df_balanced.drop('is_fraud', axis=1)
y = df_balanced['is_fraud']


rf = RandomForestClassifier()
rf.fit(X, y)

# Display feature importance
feature_importance = pd.Series(rf.feature_importances_, index=X.columns)
print(feature_importance)
```

# 2. Model

### Import Libaries

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

### Split dataset


```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Train the Model

```python
rf = RandomForestClassifier(random_state=42, class_weight='balanced')
rf.fit(X_train, y_train)
```

### Make predictions

```python
y_pred = rf.predict(X_test)
```

### Evaluate the model

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Generate classification report
class_report = classification_report(y_test, y_pred)

# Print evaluation metrics
print(f'Accuracy: {accuracy}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```
